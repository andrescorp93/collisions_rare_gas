```python
import numpy as np
import matplotlib.pyplot as plt
print('Enter file name of ground state')
fn_1s5 = input()
print('Enter file name of excited state')
fn_2p9 = input()
en1s5_file = open(fn_1s5, "r")
en2p9_file = open(fn_2p9, "r")
m1s5 = [line.strip() for line in en1s5_file.readlines()]
m2p9 = [line.strip() for line in en2p9_file.readlines()]
en1s5_file.close()
en2p9_file.close()
keys_1s5 = m1s5[0].split()
data_1s5 = {}
keys_2p9 = m2p9[0].split()
data_2p9 = {}
for j in range(len(keys_1s5)):
    temp_array = np.array([])
    for i in range(1,len(m1s5)):
        temp_str = m1s5[i].split()
        temp_array = np.append(temp_array,float(temp_str[j]))
    data_1s5[keys_1s5[j]]=temp_array
    #if j > 0: plt.plot(data_1s5[keys_1s5[0]],data_1s5[keys_1s5[j]])
for j in range(len(keys_2p9)):
    temp_array = np.array([])
    for i in range(1,len(m2p9)):
        temp_str = m2p9[i].split()
        temp_array = np.append(temp_array,float(temp_str[j]))
    data_2p9[keys_2p9[j]]=temp_array
    #if j > 0: plt.plot(data_2p9[keys_2p9[0]],data_2p9[keys_2p9[j]])
a1 = 'A1'
b1 = 'B1'
b2 = 'B2'
a2 = 'A2'
rule1 = lambda keyup, keydown: ((keyup.find(a1) != -1) and (keydown.find(b1) != -1)) or ((keyup.find(b1) != -1) and (keydown.find(a1) != -1))
rule2 = lambda keyup, keydown: ((keyup.find(b2) != -1) and (keydown.find(a2) != -1)) or ((keyup.find(a2) != -1) and (keydown.find(b2) != -1))
rule3 = lambda keyup, keydown: ((keyup.find(a1) != -1) and (keydown.find(b2) != -1)) or ((keyup.find(b2) != -1) and (keydown.find(a1) != -1))
rule4 = lambda keyup, keydown: ((keyup.find(b1) != -1) and (keydown.find(a2) != -1)) or ((keyup.find(a2) != -1) and (keydown.find(b1) != -1))
rule5 = lambda keyup, keydown, sym: ((keyup.find(sym) != -1) and (keydown.find(sym) != -1))
rulex = lambda keyup, keydown: rule1(kup,kdown) or rule2(kup,kdown)
ruley = lambda keyup, keydown: rule3(kup,kdown) or rule4(kup,kdown)
rulez = lambda keyup, keydown: rule5(kup,kdown,a1) or rule5(kup,kdown,b1) or rule5(kup,kdown,b2) or rule5(kup,kdown,a2)
trans = {}
weights = {}
for kup in keys_2p9:
    for kdown in keys_1s5:
        if rulex(kup,kdown) or ruley(kup,kdown) or rulez(kup,kdown):
            k = kup + "_" + kdown
            trans[k] = data_2p9[kup] - data_1s5[kdown]            
trans['R'] = data_2p9[keys_2p9[0]]
#print(trans)
dzetas = {}
dzetab = {}
T = np.arange(100,500,10)
v = 8.28e3*np.sqrt(T)
c = 2.998e2
dzetas['T'] = T
dzetab['T'] = T
print('Enter file name of shifts file')
outsname = input()
print('Enter file name of broadening file')
outbname = input()
open(outsname, "w").close()
dzetasf = open(outsname, "w")
open(outbname, "w").close()
dzetabf = open(outbname, "w")
for t in trans.keys():
    r = trans["R"]
    sigb = np.zeros(len(T))
    sigs = np.zeros(len(T))
    if t!="R":
        w = trans[t]
        dw = w - w[0]*np.ones(len(w))
        eta = np.zeros(len(r))
        for i in range(2,len(r)-1):
            for j in range(1,i-1):
                eta[i] += np.sqrt(r[j]**2-r[i]**2)*(dw[j-1]-dw[j+1])
        for j in range(len(T)):
            for i in range(len(eta)):
                sigb[j] += 2*np.pi*(1-np.cos(eta[i]*2*c/v[j]))*r[i]*(-r[i]+r[i-1])*1e-16
                sigs[j] += 2*np.pi*np.sin(eta[i]*2*c/v[j])*r[i]*(-r[i]+r[i-1])*1e-16
        dzetab[t]=sigb*v
        dzetas[t]=sigs*v
for k in dzetas.keys():
    dzetasf.write(k+"\t")
dzetasf.write("\n")
for j in range(len(T)):
    for k in dzetas.keys():
        if k == "T":
            dzetasf.write(str(dzetas[k][j])+"\t")
        else:
            dzetasf.write("{0:.4}\t".format(dzetas[k][j]))
    dzetasf.write("\n")
dzetasf.close()
for k in dzetab.keys():
    dzetabf.write(k+"\t")
dzetabf.write("\n")
for j in range(len(T)):
    for k in dzetab.keys():
        if k == "T":
            dzetabf.write(str(dzetab[k][j])+"\t")
        else:
            dzetabf.write("{0:.4}\t".format(dzetab[k][j]))
    dzetabf.write("\n")
dzetabf.close()
#plt.plot(T,dzeta[t])
#plt.show()
#print(dzeta)
#print(sigmas)
#tot_dzeta = 0
#for k in dzeta.keys():
#    tot_dzeta += dzeta[k]
#tot_dzeta /= 27
#print(tot_dzeta)
```
